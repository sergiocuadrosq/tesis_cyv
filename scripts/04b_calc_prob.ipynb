```python
import joblib
import pandas as pd
from itertools import combinations

# Cargar los objetos guardados
logit = joblib.load("../output/modelos/modelo_logit.pkl")
scaler = joblib.load("../output/modelos/scaler.pkl")
selected_features = joblib.load("../output/modelos/features.pkl")
```

```python
df_linea_final = pd.read_csv(r'../data/final/2_trasversal/3_modelling/df_linea_final.csv')
df_linea_final
print(df_linea_final.info())
```

```python
# Predicción de clase (0 o 1)
y_pred = logit.predict(df_linea_final)

# Probabilidad de que sea 1
y_prob = logit.predict_proba(df_linea_final)[:, 1]

# Combinar resultados en un DataFrame
resultados = pd.DataFrame({
    "prediccion": y_pred,
    "probabilidad": y_prob
})

print(resultados.head())
```

```python
print(y_prob.mean(), y_prob.std())
```

```python
df_linea_final['target'] = y_pred
print(df_linea_final['target'].value_counts(normalize=True))
```

```python
# Seleccionar la columna 'target' de df_linea_final
target_column = df_linea_final['target']

# Verificar el índice de la columna 'target'
print(target_column.index)
```

```python
# Importar el CSV con los datos originales (features)
df_features_original = pd.read_csv(r'../data/final/2_trasversal/4_linea/df_features_original.csv')

# Verifica las primeras filas para asegurarte de que se importó correctamente
print(df_features_original.head())
```

```python
# Asegurarse de que ambos DataFrames tienen el mismo índice
df_features_original = df_features_original.reset_index(drop=True)

# Unir la columna 'target' con 'df_features_original'
df_predicted_lineavulnerabilidad = pd.concat([df_features_original, target_column], axis=1)

# Verificar las primeras filas del DataFrame resultante
print(df_predicted_lineavulnerabilidad.head())
```

```python
# Guardar el DataFrame combinado en un archivo CSV
df_predicted_lineavulnerabilidad.to_csv(r'../data/final/2_trasversal/4_linea/df_predicted_lineavulnerabilidad.csv', index=False)

# Confirmación de que el archivo se guardó correctamente
print("✅ Nuevo dataset guardado como df_predicted_lineavulnerabilidad.csv")
```
